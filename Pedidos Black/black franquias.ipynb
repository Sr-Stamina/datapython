```python
import pandas as pd

# Define the required quantities per order
data_required = [
    {'Grupo': 'Masculino', 'Linha': 'Camiseta', 'P':30, 'M':60, 'G':40, 'GG':30, 'XG':2, '2XG':2, '3XG':2, 'Total':166, 'Preço Custo (R$)':39.9, 'Preço Venda (R$)':89.9},
    {'Grupo': 'Masculino', 'Linha': 'Polo', 'P':10, 'M':20, 'G':20, 'GG':10, 'XG':3, '2XG':2, '3XG':1, 'Total':66, 'Preço Custo (R$)':59.9, 'Preço Venda (R$)':129.9},
    {'Grupo': 'Masculino', 'Linha': 'CAMISA LONGA', 'P':3, 'M':5, 'G':4, 'GG':3, 'XG':2, '2XG':2, '3XG':2, 'Total':21, 'Preço Custo (R$)':69.9, 'Preço Venda (R$)':149.9},
    {'Grupo': 'Masculino', 'Linha': 'CAMISA CURTA', 'P':2, 'M':5, 'G':5, 'GG':2, 'XG':2, '2XG':2, '3XG':2, 'Total':20, 'Preço Custo (R$)':69.9, 'Preço Venda (R$)':149.9},
    {'Grupo': 'Masculino', 'Linha': 'Bermuda', 'P':2, 'M':4, 'G':2, 'GG':2, 'Total':10, 'Preço Custo (R$)':39.9, 'Preço Venda (R$)':89.9},
    {'Grupo': 'Masculino', 'Linha': 'BONE', 'UN':40, 'Total':40, 'Preço Custo (R$)':35.9, 'Preço Venda (R$)':79.9},
    
    {'Grupo': 'Feminino', 'Linha': 'Camiseta', 'P':20, 'M':15, 'G':15, 'GG':15, 'Total':65, 'Preço Custo (R$)':39.9, 'Preço Venda (R$)':89.9},
    {'Grupo': 'Feminino', 'Linha': 'Polo', 'P':5, 'M':5, 'G':5, 'GG':5, 'Total':20, 'Preço Custo (R$)':49.9, 'Preço Venda (R$)':119.9},
    {'Grupo': 'Feminino', 'Linha': 'Bermuda', 'P':4, 'M':4, 'G':4, 'GG':4, 'Total':16, 'Preço Custo (R$)':39.9, 'Preço Venda (R$)':89.9},
    {'Grupo': 'Feminino', 'Linha': 'Camisa Longa', 'P':4, 'M':3, 'G':3, 'GG':3, 'Total':13, 'Preço Custo (R$)':64.9, 'Preço Venda (R$)':139.9},
    {'Grupo': 'Feminino', 'Linha': 'Cropped', 'P':5, 'M':5, 'G':5, 'GG':5, 'Total':20, 'Preço Custo (R$)':34.9, 'Preço Venda (R$)':79.9},
    {'Grupo': 'Feminino', 'Linha': 'BONE', 'UN':10, 'Total':10, 'Preço Custo (R$)':35.9, 'Preço Venda (R$)':79.9},

    {'Grupo': 'INF MASCULINO', 'Linha': 'CAMISETA', '2':5, '4':5, '6':5, '8':5, '10':5, '12':5, 'Total':30, 'Preço Custo (R$)':29.9, 'Preço Venda (R$)':69.9},
    {'Grupo': 'INF MASCULINO', 'Linha': 'POLO', '2':2, '4':2, '6':4, '8':2, '10':2, '12':2, 'Total':14, 'Preço Custo (R$)':39.9, 'Preço Venda (R$)':99.9},
    {'Grupo': 'INF MASCULINO', 'Linha': 'CAMISA LONGA', '2':2, '4':2, '6':4, '8':2, '10':2, '12':2, 'Total':14, 'Preço Custo (R$)':69.9, 'Preço Venda (R$)':109.9},
    {'Grupo': 'INF MASCULINO', 'Linha': 'CAMISA CURTA', '2':2, '4':2, '6':4, '8':2, '10':2, '12':2, 'Total':14, 'Preço Custo (R$)':39.9, 'Preço Venda (R$)':99.9},

    {'Grupo': 'CALCADOS MASC', 'Linha': 'CHINELO', '37':2, '38':2, '39':2, '40':4, '41':2, '42':2, '43':2, 'Total':12, 'Preço Custo (R$)':24.9, 'Preço Venda (R$)':59.9},
]

size_columns = ['P', 'M', 'G', 'GG', 'XG', '2XG', '3XG', '37', '39', '40', '41', '42', '43', '2', '4', '6', '8', '10', '12', 'UN']

df_required = pd.DataFrame(data_required)

# Read the stock data from your Excel file
stock_file_path = 'C:/Users/jhonathan.alves/Documents/Python scripts/produtos em estoque.xlsx'
df_stock = pd.read_excel(stock_file_path)

# Ensure consistent data formats
df_stock['cod_tamanho'] = df_stock['cod_tamanho'].astype(str).str.upper()
df_stock['desc_linha'] = df_stock['desc_linha'].str.upper()
df_stock['desc_grupo'] = df_stock['desc_grupo'].str.upper()
df_stock['referencia'] = df_stock['referencia'].astype(str)

df_required['Grupo'] = df_required['Grupo'].str.upper()
df_required['Linha'] = df_required['Linha'].str.upper()
size_columns = [size.upper() for size in size_columns]

num_orders = 39  # Number of orders to create
orders = []
used_references = set()

# Create a copy of the stock DataFrame to avoid modifying the original
df_stock_available = df_stock.copy()

for order_num in range(num_orders):
    print(f"Processing order {order_num+1}")
    order_products = []
    order_used_references = set()

    for idx, row in df_required.iterrows():
        grupo = row['Grupo']
        linha = row['Linha']

        for size in size_columns:
            if size in row and not pd.isna(row[size]) and row[size] > 0:
                required_qty = int(row[size])

                while required_qty > 0:
                    # First, try to find products with unused references
                    matching_products = df_stock_available[
                        (df_stock_available['desc_grupo'] == grupo) &
                        (df_stock_available['desc_linha'] == linha) &
                        (df_stock_available['cod_tamanho'] == size) &
                        (df_stock_available['qtde_est'] > 0) &
                        (~df_stock_available['referencia'].isin(used_references)) &
                        (~df_stock_available['referencia'].isin(order_used_references))
                    ]

                    if matching_products.empty:
                        # If none, allow used references if no other references are available
                        matching_products = df_stock_available[
                            (df_stock_available['desc_grupo'] == grupo) &
                            (df_stock_available['desc_linha'] == linha) &
                            (df_stock_available['cod_tamanho'] == size) &
                            (df_stock_available['qtde_est'] > 0) &
                            (~df_stock_available['referencia'].isin(order_used_references))
                        ]

                    if matching_products.empty:
                        print(f"Not enough stock to fulfill size {size} for {grupo} {linha} in order {order_num+1}")
                        break
                    else:
                        # Get the product with the highest stock to maximize allocation
                        product = matching_products.sort_values(by='qtde_est', ascending=False).iloc[0]
                        # Add the product to the order
                        order_products.append(product)
                        required_qty -= 1
                        product_index = product.name
                        # Decrease stock quantity
                        df_stock_available.at[product_index, 'qtde_est'] -= 1
                        # If stock reaches zero, mark reference as used globally
                        if df_stock_available.at[product_index, 'qtde_est'] == 0:
                            used_references.add(product['referencia'])
                        # Mark reference as used in the current order
                        order_used_references.add(product['referencia'])

    # Append the DataFrame of products for the current order
    order_df = pd.DataFrame(order_products)
    # Since products may have been added multiple times, we need to count quantities
    order_summary = order_df.groupby(order_df.columns.tolist()).size().reset_index(name='Quantity')
    orders.append(order_summary)

# Save each order to a separate sheet in an Excel workbook
with pd.ExcelWriter('C:/Users/jhonathan.alves/Documents/Python scripts/Pedidos.xlsx') as writer:
    for idx, order_df in enumerate(orders):
        sheet_name = f'Pedido_{idx+1}'
        order_df.to_excel(writer, sheet_name=sheet_name, index=False)

print("Orders have been saved to Pedidos.xlsx")
```
