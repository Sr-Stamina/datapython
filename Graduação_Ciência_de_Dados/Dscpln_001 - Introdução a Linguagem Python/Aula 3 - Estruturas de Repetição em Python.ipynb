> Repetição em `for`. O método `for` permite percorrer uma lista de valores

```python
# Lista simples - MANUAL
lojas = ['LOJA 0 - MATRIX', 'LOJA 1', 'LOJA 2', 'LOJA 3']
```

#### Sintaxe do `range`.
- Com um valor `range[10]`, vai percorrer os 10 primeiros índices
- Com dois valores `range[5, 25]`, vai percorrer do 5° índice ao 25° índice
- Com três valores `range[0, 50, 5]`, vai percorrer até o 50° índice, pulando de 5 em 5
> Os valores devem ser separados por vírgula

```python
# Lista com FOR e passando um RANGE
lista_lojas_for = []
for i in range(10000):
    lista_lojas_for.append(f'LOJA {i}')

print(lista_lojas_for)
```

> Estrutura de repetição `while`. A estrutura `while` permite controle do loop, por meio de condições
---
- O `while` é usado quando não se sabe onde vai parar a repetição, mas deseja parar quando uma condição for atendida

```python
num = int(input('Digite um número (ou 0 para sair): '))
while num != 0:
    if num % 2 == 0:
        print('O número é par.')
    else:
        print('O número é impar.')
    num = int(input('Digite um número (ou 0 para sair): '))
```

> Contro de repetição `range`, `break` e `continue`. Também são usadas para repetição.
