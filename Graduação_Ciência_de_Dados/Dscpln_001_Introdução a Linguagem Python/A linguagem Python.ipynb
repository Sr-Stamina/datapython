#**Variáveis e tipos de dados**

entrada -> processamento -> saída

##Tipos de variáveis

Inteiro ou int (<class 'int'>) -> 1, 2, 3

Texto ou string (<class 'str'>) -> apple, maçã, violão, uva

Real ou float (<class 'float'>) -> 5.1, 12.97, 15.43485618

Booleano ou bool (<class 'bool'>) -> true, false

```python
med = (nota_00 + nota_01 + nota_02 + nota_03 + nota_04 + nota_05) / 6
```

```python
s = ['aprovado', 'recuperação', 'reprovado']

if med >= 5.5:
  cond = s[0]
elif med >= 4.5:
  cond = s[1]
else:
  cond = s[2]
```

```python
if s == 'aprovado':
  print(f'A sua média é: {med} e você está {s[0]}')
elif s == 'recuperação':
  print(f'A sua média é: {med} e infelizmente você esta em {s[1]}')
else:
  s == 'reprovado'
  print(f'Infelizmente você está {s[2]} e sua média é de {med}')
```
