# Elementos do M.E.R. - **Modelo Entidade Relacionamento**
---

### Relacionamentos
> O relacionamento é a associação entre entidades
- Sendo que cada uma dessas entidades é responsável por desempenhar uma função dentro desse relacionamento.
- Exemplo: Imagine as entidades `Médico`, `Paciente` e `Hospital` em um universo.
- Um possível relacionamento entre elas é o `Atendimento`, de modo que um médico possa atender um paciente em um determinado hospital.

---

### Atributos
> Os **atributos** são as características específicas de uma entidade
- Além disso, há também a possibilidade de haver **atributos** nos relacionamentos
- Exemplo, na entidade `Pessoa` pode haver os **atributos** `nome`, `CPF`, `telefone`, `endereço`, entre outros elementos que possam descrever as características de um indivíduo

- Outro exemplo: No relacionamento `Casamento` pode haver os **atributos** `Data` e `Local`.
